```python
import geopandas as gp
import rasterio
from osgeo import gdal
```

```python
raster_path = "/Users/propoop/Downloads/hennepin extreme heat mao/env_cva_lst2022.gdb"

gdal_dataset = gdal.Open(raster_path)
print(gdal_dataset)  # This might be None unless you pass a subdataset

# Try listing subdatasets:
ds = gdal.OpenEx(raster_path, gdal.OF_READONLY | gdal.OF_RASTER)
ds.GetSubDatasets()
```

```python
import matplotlib.pyplot as plt
import numpy as np

with rasterio.open(raster_path) as src:
    # Read only a small window
    window = rasterio.windows.Window(0, 0, 100, 100)
    data = src.read(1, window=window)
    
    plt.imshow(data, cmap='inferno')
    plt.colorbar(label="Surface Temperature")
    plt.title("Sample Window of LST Raster")
    plt.show()
```
